**10.Plot following  graph using data given in 3 (observe 2x2 subplots without 
labels)**

```
import numpy as np
import matplotlib.pyplot as plt

cgpa1=[9.2,6.2,8.6,8.2,6.7,5.4,6,8] 
cgpa2=[7,7.7,6.7,9,8,5.6,8,6]
cgpa3=[7.2,6.7,7.7,7,6,6.6,7,5.4] 
cgpa4=[7.6,8.8,8.7,9,8,7.6,9,9.4]
sem=[1,2,3,4,5,6,7,8]

fig, axes = plt.subplots(2,2,figsize=(11,6))

# first subplot 
axes[0,0].plot(sem, cgpa1)
axes[0,0].set_title("CGPA 1")

# second subplot
axes[0,1].plot(sem, cgpa2)
axes[0,1].set_title("CGPA 2")
axes[0,1].set_yticks([6, 7, 8, 9]) 

# third subplot 
axes[1,0].bar(sem, cgpa3, width=0.6)
axes[1,0].set_xlim(0, 8.5)                # give margin so last bar fits
axes[1,0].set_xticks([0,2,4,6,8])
axes[1,0].set_title("CGPA 3")

# fourth subplot
axes[1,1].bar(sem, cgpa4, width=0.3)      # use sem, not np.arange
axes[1,1].set_xlim(0, 8.5)
axes[1,1].set_xticks([0,2,4,6,8])
axes[1,1].set_title("CGPA 4")

plt.tight_layout()
plt.show()
```
